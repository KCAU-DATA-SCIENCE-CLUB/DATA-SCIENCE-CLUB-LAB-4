```python
# Function to identify outliers using IQR
def identify_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]


# Identify outliers in specific columns
outliers_spec_score = identify_outliers_iqr(df, 'Rating')
outliers_likes = identify_outliers_iqr(df, 'Specs Score')
outliers_ratings = identify_outliers_iqr(df, 'Likes')
```

```python
# Create a copy of the original DataFrame to preserve the original data
df_cleaned = df.copy()

# Remove rows with outliers in the 'Specs Score' column
df_cleaned = df_cleaned[~df_cleaned['Specs Score'].isin(outliers_spec_score['Specs Score'])]

# Remove rows with outliers in the 'Likes' column
df_cleaned = df_cleaned[~df_cleaned['Likes'].isin(outliers_likes['Likes'])]

# Remove rows with outliers in the 'Rating' column
df_cleaned = df_cleaned[~df_cleaned['Rating'].isin(outliers_ratings['Rating'])]
```

# Creating New columns

```python
# Extract Phone Brand, Model, and Series
df.loc[:, 'Phone Brand'] = df['Phone Title'].str.split().str[0]
df.loc[:, 'Phone Model'] = df['Phone Title'].str.split().str[1]
df.loc[:, 'Phone Series'] = df['Phone Title'].str.split().str[2]

# Extract specifications from the Specs column
df.loc[:, 'Screen'] = df['Specs'].str.split('Front Camera:').str[0].str.replace('Screen:', '', regex=False)
df.loc[:, 'Front Camera'] = df['Specs'].str.split('OS:').str[0].str.split('Front Camera:').str[1]
df.loc[:, 'OS'] = df['Specs'].str.split('Storage:').str[0].str.split('OS:').str[1]
df.loc[:, 'Storage'] = df['Specs'].str.split('Battery:').str[0].str.split('Storage:').str[1]
df.loc[:, 'Battery'] = df['Specs'].str.split('Battery:').str[1]
df
```

# Finding the most popular phone

```python
# Extracting the brand from the Phone Title 
df.loc[:, 'Brand'] = df['Phone Title'].str.split().str[0]
```

```python
# Group by brand and sum the Likes
brand_likes = df.groupby('Brand')['Likes'].sum()
```

```python
# Display the result
print(f"The most popular phone brand is '{most_popular_brand}' with a total of {most_popular_likes} likes.")
```
