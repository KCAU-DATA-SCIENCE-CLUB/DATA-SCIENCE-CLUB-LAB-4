```python
import pandas as pd
```

```python
df = pd.read_csv("PhonesPriceInKenya.csv")
```

```python
df
```

```python
#checking for null values
df.isna().sum()
```

```python
#droping null values in the Price(Kshs) column
df = df.dropna(subset=['Price(Kshs)'])
df.isna().sum()
```

```python
#creating columns
#screen inches
df['screen_inches']= df['Specs'].str.split(' ').str[0]
df
```

```python
df['screen_inches1']= df['screen_inches'].str.split(':').str[1]
df
```

```python
#creating a pixels column
df['pixels']= df['Specs'].str.split(' ').str[2:5]
df
```

```python
#creating a column . phone model
df['phone_model']= df['Phone Title'].str.split(' ').str[0:2]
df
```

```python
#Find the top 10 most expensive Samsung phones
#creating a column for samsung phones only
df['Samsung_Phones'] = df['Phone Title'].apply(lambda x: x if 'Samsung' in x else None )
df
```

```python
#creating a dataframe for samsung phones only
prices_columns=['Samsung_Phones','Price(Kshs)']
prices_columns_df = df[prices_columns]
prices_columns_df
```

```python
#remoing null values from the prices_columns_df dataframe
prices_columns_df= prices_columns_df.dropna()
prices_columns_df
```

```python
# Get the 10 most expensive phones in the dataframe prices_columns_df
# Remove commas from the 'Price(Kshs)' column and convert to float

# Ensure all values in 'Price(Kshs)' column are strings before using .str accessor
prices_columns_df['Price(Kshs)'] = prices_columns_df['Price(Kshs)'].astype(str)
prices_columns_df['Price(Kshs)'] = prices_columns_df['Price(Kshs)'].str.replace(',', '').astype(float)

# Find the 10 most expensive phones
top_10_expensive_phones = prices_columns_df.nlargest(10, 'Price(Kshs)')
top_10_expensive_phones
```

```python
#Find the top 10 most cheapest apple phones
#creating a column for apple phones only
df['Apple iPhone'] = df['Phone Title'].apply(lambda x: x if 'Apple iPhone' in x else None )
df
```

```python
# Creating a dataframe for apple phones only
prices_columns_apple = ['Apple iPhone', 'Price(Kshs)']
prices_columns_apple_df = df[prices_columns_apple]
prices_columns_apple_df
```

```python
# Ensure all values in 'Price(Kshs)' column are strings before using .str accessor
prices_columns_apple_df['Price(Kshs)'] = prices_columns_apple_df['Price(Kshs)'].astype(str)
prices_columns_apple_df['Price(Kshs)'] = prices_columns_apple_df['Price(Kshs)'].str.replace(',', '').astype(float)

# Removing null values from the prices_columns_apple_df dataframe

prices_columns_apple_df = prices_columns_apple_df.dropna()
prices_columns_apple_df
```

```python
# finding the cheapest iphones
cheapest_iphones = prices_columns_apple_df.nsmallest(10, 'Price(Kshs)')
cheapest_iphones
```

```python
#How many Samsung Galaxy A series are in the dataset?
#creating a column for galaxy A series phones

df['Samsung Galaxy A'] = df['Phone Title'].apply(lambda x: x if 'Samsung Galaxy A' in x else None )
df
```

```python
#finding the most popular phone brand
# Extracting the phone brand from the 'Phone Title' column
df['Brand'] = df['Phone Title'].apply(lambda x: x.split()[0])

# Counting the number of occurrences of each brand
brand_counts = df['Brand'].value_counts()

# Getting the most popular phone brand
most_popular_brand = brand_counts.idxmax()
most_popular_brand
```

```python
#Does likes affect the ratings of a phone model?
# Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Checking the relationship between 'Likes' and 'Rating'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Likes', y='Rating')
plt.title('Relationship between Likes and Rating')
plt.xlabel('Likes')
plt.ylabel('Rating')
plt.show()

# Calculating the correlation coefficient between 'Likes' and 'Rating'
correlation_likes_rating = df['Likes'].corr(df['Rating'])
correlation_likes_rating
```

```python
# Does ratings affect the price of a phone model?
# Checking the relationship between 'Rating' and 'Price(Kshs)'
sample_df = df.sample(n=30, random_state=1)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=sample_df, x='Rating', y='Price(Kshs)')
plt.title('Relationship between Rating and Price')
plt.xlabel('Rating')
plt.ylabel('Price (Kshs)')
plt.show()



# Remove commas from the 'Price(Kshs)' column and convert to float
# Ensure all values in 'Price(Kshs)' column are strings before using .str accessor
df['Price(Kshs)'] = df['Price(Kshs)'].astype(str)
df['Price(Kshs)'] = df['Price(Kshs)'].str.replace(',', '').astype(float)


# Calculating the correlation coefficient between 'Rating' and 'Price(Kshs)'
correlation_rating_price = df['Rating'].corr(df['Price(Kshs)'])
correlation_rating_price
```
